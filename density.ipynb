### Average and range

```python
def data_plot(n):
    value = 0
    cont = 0
    AVERAGE = []
    RANGE = []
    range_tem = []
    for row in df['DENSIDAD OPTICA']:
        #print(row)
        cont +=1 
        value = value + row
        range_tem.append(row)
        if cont == n: # this checks if the size of the sample is equal to n
            AVERAGE.append(round((value/n),4))
            cont = 0
            value = 0
            RANGE.append(max(range_tem) - min(range_tem))
            range_tem.clear()
        else:
            continue

    average_density = sum(AVERAGE)/len(AVERAGE)
    average_range = sum(RANGE)/len(RANGE) 
    return [AVERAGE, average_density,average_range]

    
    
    
```

### Histogram

```python
denisity_list = df['DENSIDAD OPTICA'].values.tolist()

#kwargs = dict(alpha=0.5, bins=100, density=True, stacked=True)
plt.figure(figsize=(30, 20))
plt.ylabel('Frequency',fontsize=20)
plt.xlabel('Density',fontsize=20)
plt.grid(True)
plt.hist(denisity_list, color='g', bins=100,alpha=0.5,width=0.009)
```

```python
dias = ['Mon 1st','Mon 2nd','Tue 1st','Tue 2nd','Wed 1st','Wed 2nd','Thu 1st','Thu 2nd','Fri 1st','Fri 2nd','Sat 1st','Sat 2nd','Sun 1st','Sun 2nd']
dias = dias*10
mapeado = range(len(dias))

plt.figure(figsize=(30, 10))
plt.suptitle(f'Average density Sample = 5',fontsize=25)
plt.ylabel('Density',fontsize=20)
plt.xlabel('Day of the week',fontsize=20)
plt.xticks(mapeado,dias,rotation=90,fontsize=10)
plt.grid(True)

AVERAGE = data_plot(5)[0]

plt.plot(AVERAGE)

plt.axhline(y=1.125, color='green', linewidth=5,alpha=0.7)
plt.axhline(y=1.199, color='red',linewidth=5,alpha=0.4)
plt.axhline(y=1.051, color='red',linewidth=5,alpha=0.4)
plt.axhline(y=1.175 ,color='orange',linewidth=3,linestyle='dashed',alpha=0.5)
plt.axhline(y=1.150, color='blue',linewidth=3,linestyle='dashed',alpha=0.5)
plt.axhline(y=1.076, color='orange',linewidth=3,linestyle='dashed',alpha=0.5)
plt.axhline(y=1.101, color='blue',linewidth=3,linestyle='dashed',alpha=0.5)
plt.show()
```
